## 测试实验
关于编队的仿射变换在手内旋转任务学习中的效果，并与关节空间进行对比。
主要的思想就是预设一个手构型 $P^*$, RL学的是仿射变换矩阵 $A$ 和平移向量 $b$，作用于手指末端位置{b}并不断更新

$$P=(I_n \otimes A)P^*+1_n \otimes b$$
$P \in \mathbb{R}^{12}$: 末端的3D坐标。因为有4根手指，每根手指3个坐标，所以是12维
<br>$A \in \mathbb{R}^{3 \times 3}$: 仿射变换矩阵
<br>$P^* \in \mathbb{R}^{12}$: 标称末端位置(编队里的Nominal Configuration)
<br>$b \in \mathbb{R}^{3}$: 平移向量

关于工程实现的几个注意事项
1. 坐标系的选定，目前偏西{e},在各自的环境坐标系并行训练
2. 仿射矩阵 $A$ 如何学习，是分解成 $R$ 和 $D$，还是直接学9个分量？
3. 逆运动学，因为这里只更新的位置，故求解时优先满足位置的要求，打算采用 `DifferentialIKController`
